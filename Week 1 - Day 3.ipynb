```python
numbers=[10, 20, 30]

print(numbers)
print(numbers[0])
print(numbers[1])
```

```python
numbers.append(40)
print(numbers)
```

```python
for n in numbers:
    print(n)
```

```python
student={
    "name":"Jimin",
    "age":22,
    "major":"Lojistics"
}

print(student)
```

```python
print(student["name"])
print(student["age"])
```

```python
for key in student:
    print(key, student[key])
```
