```python
import pandas as pd

df = pd.read_csv("data.csv.txt")
df
```

```python
df["result"] = df["score"] >= 70
df
```

```python
df["grade"] = df["score"].apply(
    lambda x: "PASS" if x >= 70 else "FAIL"
)
df
```

```python
df_sorted= df.sort_values(by="score", ascending=False)
df_sorted
```
